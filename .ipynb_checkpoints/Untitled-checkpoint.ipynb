```python
from datetime import datetime

start = datetime(2018, 1, 1)
end = datetime(2019, 2, 28)

symbols = ['EUR/USD', 'USD/JPY']
df = dict()
for symbol in symbols:
    df[symbol] = con.get_candles(symbol, period='H1', number=110)['askclose']
```

```python
import pandas as pd
fx = pd.DataFrame(df)
fx.head()
```

```python
# รีเทิร์น เทียบกับจุดแรกสุด ให้จุดแรกสุดเป็นหนึ่ง
stock_return = fx.apply(lambda x: x / x[0])
stock_return.plot(grid = True)
```

```python
import numpy as np
pct_return = fx.pct_change()
```

```python
pct_return.plot()
```

```python
# There are 252 trading days in a year; the 100 converts to percentages
# annual percentage rate (APR)
fx_change_apr = pct_return * 252 * 100    
fx_change_apr.tail()
```

```python
eu = pd.DataFrame(df['EUR/USD'])
```

```python
eu['20'] = eu['askclose'].rolling(window=20, center=False).mean()
eu['50'] = eu['askclose'].rolling(window=20, center=False).mean()
eu['20-50'] = eu['20'] - eu['50']
eu['Regime'] = np.where(eu['20-50'] > 0, 1, 0)
```

```python
eu.tail()
```

```python
def z(df):
    return (df - df.mean()) / df.std()
```

```python
eu['z'] = z(eu['askclose'])
eu['z_20'] = eu['z'].rolling(20).mean()
eu['z_50'] = eu['z'].rolling(50).mean()
eu['z_20-50'] = eu['z_20'] - eu['z_50']
eu['z_regime'] = np.where(eu['z_20-50'] > 0, 1, 0)
eu['z_regime'] = np.where(eu['z_20-50'] < 0, -1, eu['z_regime'])
eu['z_regime'].plot()
```

```python
eu['z_regime'].value_counts()
```
