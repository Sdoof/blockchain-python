# Calculation formulas

### Margin
Lots x Contract size / Leverage

1.00 x 100000.0 / 10 = 10000.000 NZD

### Swap
Lots x Contract size x Short_or_Long x Point size

1.00 x 100000.0 x -0.21300 x 0.00010 = -2.13

1.00 x 100000.0 x 0.06500 x 0.00010 = 0.65

### Point profit
Point size x Contract size x Lots

0.00010 x 100000.0 x 1.00 = 10.00 USD

### Volume
Lots x Contract size

1.00 x 100000 = 100000.00 NZD

### Volatility
volatility (symbol Ïƒ) is the degree of variation of a trading price series over time as measured by the standard deviation of logarithmic returns.


```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
%matplotlib inline

import fxcmpy
con = fxcmpy.fxcmpy(config_file = '/Users/ballmdr/blockchain-python/FXCM/fxcm.cfg')
```

```python
def reversionPlot(symbol):
    df = con.get_candles(symbol, period='D1', number=252)
    df['Close'] = (df['bidclose'] + df['askclose'])/2
    mean = np.mean(df['Close'])
    std = np.std(df['Close'])

    plt.title(symbol)
    plt.axvline(x=mean)
    plt.axvline(x=(mean+(std*2)))
    plt.axvline(x=(mean-(std*2)))
    plt.axvline(x=df.iloc[-1]['Close'], linestyle='--')
    sns.distplot(df['Close'])
```

```python
symbol = 'AUD/NZD'
newdf = pd.DataFrame()
df = con.get_candles(symbol, period='D1', number=252)
newdf['Open'] = (df['bidopen'] + df['askopen'])/2
newdf['High'] = (df['bidhigh'] + df['askhigh'])/2
newdf['Low'] = (df['bidlow'] + df['asklow'])/2
newdf['Close'] = (df['bidclose'] + df['askclose'])/2
newdf['return'] = newdf['Close'].pct_change()
newdf['log_return'] = np.log(1 + newdf['return'])
```

```python
newdf.tail()
```

## Money Management

```python
mean = np.mean(newdf['Close'])
high = np.max(newdf['High'])
low = np.min(newdf['Low'])
sd = np.std(newdf['Close'])
volatility = np.std(newdf['log_return'])

all_distance = high - low
contract = 1000 #0.01 lot
cash = 3000
zone = 8
bullet = 8
risk_per_zone = cash/zone
risk_per_trade = risk_per_zone/bullet
leverage = contract/risk_per_trade
zone_distance = all_distance/zone
zone_price = [0,1,2,3,4,5,6,7,8]
pipval = 0.10 #0.01 lot
last_price = newdf.iloc[-1]['Close']

plt.figure(figsize=(12,6))
plt.title(symbol)

for i in range(9):
    if i == 0:
        zone_price[0] = low
        plt.axvline(x=zone_price[0], linestyle='--')
    else:
        zone_price[i] = zone_price[i-1] + zone_distance
        plt.axvline(x=zone_price[i], linestyle='--')


plt.axvline(x=last_price, color='green')
sns.distplot(newdf['Close'])

print('Zone Distance: ' + str(zone_distance * 10000) + ' pips')
print('Risk Per Zone: ' + str(risk_per_zone))
print('Risk Per Trade: ' + str(risk_per_trade))
print('SL Distance: ' + str(risk_per_trade/pipval) + ' pips')
print('3 Standard Deviation = ' + str((sd*3)*10000))
print('3 Volatility = ' + str((volatility*3)*10000))
print('cash: ' + str(cash))
print('Leverage: ' + str(leverage))
print('bullet per zone: ' + str(bullet))
print('total bullet: ' + str(bullet * zone))
```
