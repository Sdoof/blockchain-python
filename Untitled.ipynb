The metrics I want are:

The average return
Volatility (the standard deviation of returns)
\alpha and \beta
The Sharpe ratio

```python
eu = pd.DataFrame(df['EUR/USD'])
```

```python
def z(df):
    return (df - df.mean()) / df.std()
```

```python
eu['z'] = z(eu['askclose'])
eu['z_fast'] = eu['z'].rolling(25).mean()
eu['z_slow'] = eu['z'].rolling(50).mean()
eu['z_diff'] = eu['z_fast'] - eu['z_slow']
eu['z_regime'] = np.where(eu['z_diff'] > 0, 1, 0)
eu['z_regime'] = np.where(eu['z_diff'] < 0, -1, eu['z_regime'])
eu['z_regime'].plot()
```

```python
eu['z_regime'].value_counts()
```

```python
# To ensure that all trades close out, I temporarily change the regime of the last row to 0
regime_orig = eu.loc[:, "z_regime"].iloc[-1]
eu.loc[:, "z_regime"].iloc[-1] = 0
eu["Signal"] = np.sign(eu["z_regime"] - eu["z_regime"].shift(1))
# Restore original regime data
eu.loc[:, "z_regime"].iloc[-1] = regime_orig
eu.tail()
```

```python
eu["Signal"] = np.sign(eu["z_regime"] - eu["z_regime"].shift(1))
```

```python
regime_orig = eu.loc[:, "z_regime"].iloc[-1]
regime_orig
```

```python
eu['Signal'].value_counts()
```

```python
# ทำตารางว่า บายราคาไหน เซลราคาไหน
eu_signals = pd.concat([
        pd.DataFrame({
            "Price": eu.loc[eu["Signal"] == 1, "askclose"], #เอาราคาที่มี singal = 1 มาเป็นช่อง Price
            "Regime": eu.loc[eu["Signal"] == 1, "z_regime"], #เอา regime ที่มี signal = 1 มาเป็นช่อง regime
            "Signal": "Buy" # ใส่ว่าเป็น Buy
        }),
        pd.DataFrame({
            "Price": eu.loc[eu["Signal"] == -1, "askclose"],
            "Regime": eu.loc[eu["Signal"] == -1, "z_regime"],
            "Signal": "Sell"
        }),
    ])
eu_signals.sort_index(inplace = True)
eu_signals
```

```python
# เช็คดูว่า มี buy แต่ regime ดันเป็น -1 มั้ย
eu_signals.loc[eu_signals["Signal"] == "Buy"].loc[eu_signals["Regime"] == -1]
```

```python
long_profits = pd.DataFrame({
        "Price": eu_signals.loc[eu_signals["Signal"] == "Buy", "Price"],
        "Profit": pd.Series(eu_signals["Price"] - eu_signals["Price"].shift(1)).loc[eu_signals.loc[eu_signals["Signal"].shift(1) == "Buy"].index].tolist(),
        "End Date": eu_signals["Price"].loc[eu_signals.loc[eu_signals["Signal"].shift(1) == "Buy"].index].index
    })
long_profits['Profit'].plot()
```

```python
tradeperiods = pd.DataFrame({
    "Start": long_profits.index,
    "End": long_profits["End Date"]
})

long_profits["close"] = tradeperiods.apply(lambda x: min(eu.loc[x["Start"]:x["End"], "askclose"]), axis = 1)
long_profits
```
