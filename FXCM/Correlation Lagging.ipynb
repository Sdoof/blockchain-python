```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
%matplotlib inline
import plotly
from plotly import tools
import plotly.plotly as py
import plotly.graph_objs as go
plotly.tools.set_credentials_file(username='ballmdr', api_key='oSgf6vHed7evuaiXDdBb')
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import fxcmpy
con = fxcmpy.fxcmpy(config_file = '/Users/ballmdr/blockchain-python/FXCM/fxcm.cfg')

def z(df):
    return (df - df.mean()) / df.std()
```

```python
#symbols = ['EUR/USD', 'EUR/AUD', 'EUR/CAD', 'EUR/CHF', 'EUR/JPY', 'EUR/NZD', 'EUR/GBP']
#symbols = ['GBP/AUD', 'GBP/CAD', 'GBP/CHF', 'GBP/JPY', 'GBP/NZD', 'GBP/USD']
#symbols = ['AUD/CAD', 'AUD/CHF', 'AUD/JPY', 'AUD/NZD', 'AUD/USD']
symbols = ['EUR/USD', 'USD/JPY', 'GBP/USD', 'USD/CAD', 'USD/CHF', 'AUD/USD', 'NZD/USD', 'EUR/AUD', 'EUR/CAD', 'EUR/CHF', 'EUR/GBP', 'EUR/JPY', 'EUR/NZD', 'GBP/AUD', 'GBP/CAD', 'GBP/CHF', 'GBP/JPY', 'GBP/NZD', 'AUD/CAD', 'AUD/CHF', 'AUD/JPY', 'AUD/NZD', 'NZD/CAD', 'NZD/CHF', 'NZD/JPY', 'CAD/CHF', 'CAD/JPY', 'CHF/JPY']
```

```python
df = dict()
for symbol in symbols:
    df[symbol] = con.get_candles(symbol, period='D1', number=60)
    df[symbol]['Close'] = (df[symbol]['bidclose'] + df[symbol]['askclose'])/2
    df[symbol]['Close_z'] = z(df[symbol]['Close'])
    df[symbol].dropna(inplace=True)
```

```python
plt.figure(figsize=(26,12))
for symbol in symbols:
    if df[symbol].iloc[-1]['Close_z'] > 1 or df[symbol].iloc[-1]['Close_z'] < -1:
        plt.plot(df[symbol]['Close_z'], label=symbol)
plt.legend();
```

```python
plt.figure(figsize=(20,16))
plt.plot(df['EUR/CHF']['Close_z'], label='EUR/CHF')
plt.plot(df['EUR/NZD']['Close_z'], label='EUR/NZD')
plt.legend();
```
